```python
import psycopg2
import csv

conn = psycopg2.connect(
    host="localhost",
    database="International_Debt_Analysis",
    user="postgres",
    password="Shivani1234",
    port="5432"
)

print("Connected successfully!")
```

```python
def insert_csv(file_path, table_name, query):
    try:
        cur = conn.cursor()

        with open(file_path, "r", encoding="utf-8") as f:
            reader = csv.reader(f)
            next(reader)

            for row in reader:
                cur.execute(query, row)

        conn.commit()
        print(f"{table_name} inserted successfully!")

    except Exception as e:
        conn.rollback()
        print(f"Error in {table_name}:", e)

    finally:
        cur.close()
```
